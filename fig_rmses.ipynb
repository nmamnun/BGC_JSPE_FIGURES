```python
import xarray as xr
llc90_grid = xr.open_dataset('llc90_grid.nc')
```

```python
def mask_data(data, time_dependent=False):
    # Extract the mask from the llc90_grid
    mask = llc90_grid.hFacC.values

    # Check if the mask is of the shape (50, 13, 90, 90)
    if mask.shape != (50, 13, 90, 90):
        raise ValueError("The provided grid is not an llc90 grid.")

    # Check dimensions of the data and apply the mask accordingly
    if time_dependent:
        if data.shape[1:] == (50, 13, 90, 90):
            # Extend the mask along the time dimension
            extended_mask = np.tile(mask, (data.shape[0], 1, 1, 1, 1))
            masked_data = ma.masked_where(extended_mask == 0, data)
        elif data.shape[1:] == (13, 90, 90):
            # Use only the first slice of the mask and extend along time dimension
            extended_mask = np.tile(mask[0], (data.shape[0], 1, 1, 1))
            masked_data = ma.masked_where(extended_mask == 0, data)
        else:
            raise ValueError(
                "Data dimensions with time do not match the expected shape.")
    else:
        if data.shape == (50, 13, 90, 90):
            masked_data = ma.masked_where(mask == 0, data)
        elif data.shape == (13, 90, 90):
            masked_data = ma.masked_where(mask[0] == 0, data)
        else:
            raise ValueError(
                "Data dimensions do not match the expected shape.")

    return masked_data
```

```python
cell_area = llc90_grid.rA.values
cell_area = mask_data(cell_area)
area_weight = cell_area/np.nanmean(cell_area)
# print(np.min(area_weight))
# print(np.max(area_weight))
```

```python
def calculate_rmse(predictions, observations, area):
    mask = np.isnan(observations) | predictions.mask | observations.mask
    valid_predictions = np.ma.masked_array(predictions, mask=mask)
    valid_observations = np.ma.masked_array(observations, mask=mask)
    expanded_area = np.tile(area, (predictions.shape[0], 1, 1, 1))
    valid_area = np.ma.masked_array(expanded_area, mask=mask)
    errors = (valid_predictions - valid_observations)**2
    rmse = np.sqrt(np.ma.mean(errors, axis=(1,2,3)))
    total_valid_area = np.ma.sum(valid_area, axis=(1,2,3)) 
    weighted_errors = np.ma.sum(valid_area * errors, axis=(1,2,3))
    area_weighted_rmse = np.sqrt(np.divide(weighted_errors, total_valid_area))
    return rmse, area_weighted_rmse
```

```python
obs_schla_ln_masked = mask_data(obs_schla_ln, time_dependent=True)
Surf_TotChl_freerun_ln_masked = mask_data(Surf_TotChl_freerun_ln, time_dependent=True)
Surf_TotChl_re_ana_ln_masked = mask_data(Surf_TotChl_re_ana_ln, time_dependent=True)  
Surf_TotChl_je_ana_ln_masked = mask_data(Surf_TotChl_je_ana_ln, time_dependent=True)
Surf_TotChl_je_for_ln_masked = mask_data(Surf_TotChl_je_for_ln, time_dependent=True)
```

```python
rmse_freerun, area_weighted_rmse_freerun = calculate_rmse(Surf_TotChl_freerun_ln_masked, obs_schla_ln_masked, cell_area)
rmse_re_ana, area_weighted_rmse_re_ana = calculate_rmse(Surf_TotChl_re_ana_ln_masked, obs_schla_ln_masked, cell_area)
rmse_je_for, area_weighted_rmse_je_for = calculate_rmse(Surf_TotChl_je_for_ln_masked, obs_schla_ln_masked, cell_area)
rmse_je_ana, area_weighted_rmse_je_ana = calculate_rmse(Surf_TotChl_je_ana_ln_masked, obs_schla_ln_masked, cell_area)
```
